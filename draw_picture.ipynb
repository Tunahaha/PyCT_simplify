```python
data = pd.read_csv('paper 100.csv',index_col=0)

# 定義數據
cnn_data = data[['cnn_1', 'cnn_2', 'cnn_3', 'cnn_4', 'cnn_5', 'cnn_6', 'cnn_7', 'cnn_8']]
tnn_data = data[['tnn_1', 'tnn_2', 'tnn_3', 'tnn_4', 'tnn_5', 'tnn_6', 'tnn_7', 'tnn_8']]
tnn_delta_data = data[['tnn_delta_1', 'tnn_delta_2', 'tnn_delta_3', 'tnn_delta_4', 'tnn_delta_5', 'tnn_delta_6', 'tnn_delta_7', 'tnn_delta_8']]

labels = ['1_pixel', '2_pixel', '3_pixel', '4_pixel', '5_pixel', '6_pixel', '7_pixel', '8_pixel']
x = np.arange(len(labels))

# 繪製長條圖
fig, ax = plt.subplots()
rects1 = ax.bar(x - 0.2, cnn_data.loc['example'], 0.2, label='cnn')
rects2 = ax.bar(x , tnn_data.loc['example'], 0.2, label='tnn')
rects3 = ax.bar(x + 0.2, tnn_delta_data.loc['example'], 0.2, label='tnn_delta')

# 添加一些文字標籤
ax.set_xlabel('100 Input')
ax.set_title('Total Number of Adversarial Examples')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()


```

```python
from keras.models import load_model

# 載入模型
model = load_model('model/mnist_sep_act_m6_9628.h5')

# 查看模型架構和參數數量
model.summary()


```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical


# Load MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()


# Reshape and normalize data
train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype('float32') / 255


test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype('float32') / 255


# Convert labels to categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)


# Define the model
model = tf.keras.models.Sequential([
    layers.Conv2D(2, (3, 3), padding='same', input_shape=(28, 28, 1)),
    layers.Activation('relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    
    layers.Conv2D(2, (3, 3), padding='same'),
    layers.Activation('relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),


    layers.Flatten(),
    
    layers.Dense(32),
    layers.Activation('relu'),
    
    layers.Dense(10),
    layers.Activation('softmax'),
])


# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


# Train the model
model.fit(train_images, train_labels, epochs=5, batch_size=512)


# Evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels)


# Save the model
model.save('mnist_model88.h5')



```
